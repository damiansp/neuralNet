## Recreate as TF Graph with placeholders

```python
img = tf.placeholder(tf.float32, shape = [None, None], name = 'img')
img3d = tf.expand_dims(img, 2) # adds dim to axis 2 [h, w] -> [h, w, 1]
dims = img3d.get_shape()
img4d = tf.expand_dims(img3d, 0) # [h, w, 1] -> [1, h, w, 1]
img4d.get_shape().as_list()
```

```python
# Placeholders for Gabor
mean     = tf.placeholder(tf.float32, name = 'mean')
sd       = tf.placeholder(tf.float32, name = 'sd')
kernsize = tf.placeholder(tf.int32, name = 'kernsize')
```

```python
# Convolution steps
x = tf.linspace(-3., 3., kernsize)
z = (tf.exp(tf.negative(tf.pow(x - mean, 2.) / (2. * tf.pow(sd, 2.)))) *
     (1. / (sd * tf.sqrt(2. * np.pi))))
z2d = tf.matmul(tf.reshape(z, tf.stack([kernsize, 1])),
                tf.reshape(z, tf.stack([1, kernsize])))
ys = tf.sin(x)
ys = tf.reshape(ys, tf.stack([kernsize, 1]))
ones = tf.ones(tf.stack([1, kernsize]))
wave = tf.matmul(ys, ones)
gabor = tf.multiply(wave, z2d)
gabor4d = tf.reshape(gabor, tf.stack([kernsize, kernsize, 1, 1]))
convolved = tf.nn.conv2d(img4d, 
                         gabor4d, 
                         strides = [1, 1, 1, 1], 
                         padding = 'SAME', 
                         name = 'convolved')
convolved_img = convolved[0, :, :, 0]
```

```python
res = convolved_img.eval(feed_dict = { img: data.camera(),
                                       mean: 0.,
                                       sd: 0.5,
                                       kernsize: 32 })
```

```python
plt.imshow(res, cmap = 'gray');
```
